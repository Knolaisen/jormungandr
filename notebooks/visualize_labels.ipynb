```python
from jormungandr.dataset import create_dataloaders
import torch
import matplotlib.pyplot as plt
from transformers.image_transforms import center_to_corners_format

```

```python
train_dataloader, val_dataloader = create_dataloaders(subset_size=1000)
```

```python
batch = next(iter(train_dataloader))
```

```python
print(len(batch["labels"]))
```

```python
batch_index = 5
```

```python
example_labels = batch["labels"][batch_index]
example_label_boxes = example_labels["boxes"]
example_pixel_values = batch["pixel_values"][batch_index]
example_pixel_masks = batch["pixel_mask"][batch_index]
```

```python

img = example_pixel_values[0].cpu()  # move to CPU if needed

# If shape is [C, H, W] → convert to [H, W, C]
if img.ndim == 3:

    img = img.permute(1, 2, 0)

plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
# boxes_xyxy_normalized = center_to_corners_format(example_label_boxes.cpu())
boxes_xyxy_normalized = example_label_boxes.cpu()
```

```python
boxes_xyxy_absolute = boxes_xyxy_normalized * torch.tensor([img.shape[1], img.shape[0], img.shape[1], img.shape[0]])
```

```python
boxes_xyxy_absolute
```

```python
def visualize_boxes_on_image(image, boxes):
    plt.imshow(image)
    for box in boxes:
        x_min, y_min, x_max, y_max = box
        rect = plt.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min, edgecolor='red', facecolor='none')
        plt.gca().add_patch(rect)
    plt.axis("off")
    plt.show()
visualize_boxes_on_image(img, boxes_xyxy_absolute)
```
